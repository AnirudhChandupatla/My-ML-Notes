## *Space and Time Complexity:*

 **Training Time(i.e computing distances not actual training) and Space = O( no_data_pts * dim)**
 
 This is a huge setback for this algorithm !!! as it needs high configured hardware and large amount of time.
 
 Theoretical Improvements,
*  Pre-sort training data into fast data structures(like Hash Tables)

*  Compute only an approximate distance

*  Remove redundant data (condensing)



##*Choice of Nearest Neighbors Algorithm:*

**Query time** or **prediction time** of knn for different algo's are like,

-> **Brute force** : 

         O( no_data_pts * dim )

-> **Ball tree** : 

        O( log(no_data_pts) * dim)
        
 ![alt text](https://github.com/anirudh991/My-ML-Notes/blob/master/K-Nearest-Neighbours/pictures/ball_tree.png?raw=true)

-> **kd-tree** : 

        for dim <= 20, O( log(no_data_pts) * dim) 

        else, O( no_data_pts * dim)
        
        in this case computations are done internally with dimensionality = 2^dim.
        
 ![alt text](https://github.com/anirudh991/My-ML-Notes/blob/master/K-Nearest-Neighbours/pictures/kd_tree.png?raw=true)
                            
 -> in case of **Ball tree** and **kd-tree** to get above query times **'K' must be small**
 
 -> **Brute force** performs well in case of **small no_data_pts and dim**.

![alt text](https://github.com/anirudh991/My-ML-Notes/blob/master/K-Nearest-Neighbours/pictures/effect-of-k.png?raw=true =670x500)

-> To choose right 'k' we will use cross validation

##*Weighted KNN:*

![alt text](https://github.com/anirudh991/My-ML-Notes/blob/master/K-Nearest-Neighbours/pictures/wknn.png?raw=true =310x250)

-> above for k = 3 everything is alright 

-> lets just say that we found k = 5 is suited for entire data then most likely points like that star are going to be misclassified.

-> solution is to give weights , one way is based on distances.

    to_decide_class = []
    
    while i < k:

       dist = 1/d(query-pt , i-th-pt)
            
       append {dist , class of i-th-pt} to 'to_decide_class'
    
    with this now we'll have 'k' weights(here it's based on distance) then declare the class with highest weight as class label of query-pt.
